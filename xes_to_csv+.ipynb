```python
#run 'pip install -U pm4py' in terminal
from pm4py.objects.log.importer.xes import importer as xes_importer
#from pm4py.objects.log.importer.xes import factory as xes_factory
import pandas as pd
import numpy as np
```

```python
BPI = xes_importer.apply("BPI_Challenge_2012.xes")
```

```python
listo = []
dic = {'no': -1}
prev_attr = 'no'
counter = 0
cur_attr = 'no'
for i in range(0, len(BPI)):
    for j in range(0, len(BPI[i])):
        attr_list = list(BPI[i][j])
    
        for k in range(0, len(attr_list)):
            prev_attr = cur_attr
            cur_attr = attr_list[k]    
            if cur_attr not in listo:
                value = dic[prev_attr] + 1
            
                for key in dic:
                    if dic[key] >= value:
                        dic[key] += 1
            
                dic[cur_attr] = value
                listo.insert(value, cur_attr)
    
        cur_attr = 'no'
```

```python
chain = []
event = []
for i in range(0, len(BPI)):
    for j in range(0, len(BPI[i])):
        chain.append(i)
        event.append(j)
```

```python
df_BPI= pd.DataFrame(index=[np.array(chain), np.array(event)], columns=listo)
```

```python
for i in range(0, len(BPI)):
    if i % 100 == 0:
        print(i)
    
    for j in range(0, len(BPI[i])):

        attr = BPI[i][j]
        for a in attr:
            df_BPI.loc[(i, j), a] = attr[a]
```

```python
df_BPI.to_csv('BPI.csv')
df_BPI.to_hdf('BPI.h5', 'data', mode='w')
```

```python
attr_listo = []
attr_dic = {'no': -1}
prev_attr = 'no'
counter = 0
cur_attr = 'no'
for i in range(0, len(BPI)):
    attr_list = list(BPI[i].attributes)
    
    for k in range(0, len(attr_list)):
        prev_attr = cur_attr
        cur_attr = attr_list[k]    
        if cur_attr not in attr_listo:
            value = attr_dic[prev_attr] + 1
            
            for key in attr_dic:
                if attr_dic[key] >= value:
                    attr_dic[key] += 1
            
            attr_dic[cur_attr] = value
            attr_listo.insert(value, cur_attr)
    
    cur_attr = 'no'
```

```python
attr_chain = []
for i in range(0, len(BPI)):
    attr_chain.append(i)
```

```python
df_BPI_attr= pd.DataFrame(index=[np.array(attr_chain)], columns=attr_listo)
```

```python
for i in range(0, len(BPI)):
    if i % 100 == 0:
        print(i)
    attr = BPI[i].attributes
    for a in attr:
        df_BPI_attr.loc[i, a] = attr[a]
```

```python
df_BPI_attr.to_csv('BPI_attr.csv')
df_BPI_attr.to_hdf('BPI.h5', 'data')
```
