```python
import pandas as pd
from tqdm.notebook import tqdm, trange
from IPython.core.display import display, HTML
```

```python
df = pd.read_csv('BPI_2012.csv')
```

```python
df['concept:name S1'] = df['concept:name'].shift(1)
df['concept:name S2'] = df['concept:name'].shift(2)
df['lifecycle:transition S1'] = df['lifecycle:transition'].shift(1)
df['lifecycle:transition S2'] = df['lifecycle:transition'].shift(2)
df['case_id S1'] = df['case_id'].shift(1)
df['case_id S2'] = df['case_id'].shift(2)

df[['concept:name S1','concept:name S2','lifecycle:transition S1', 'lifecycle:transition S2', 'concept:name']] = df[['concept:name S1','concept:name S2','lifecycle:transition S1', 'lifecycle:transition S2', 'concept:name']].fillna('NaN')
df[['case_id S1', 'case_id S2']] = df[['case_id S1', 'case_id S2']].fillna(0)

def absolute_concept(var):
    if (var[0][0] == 'W') and (var[1][0] == 'W') and (var[0] != var[1]) and (var[2]+var[3] == 'COMPLETESCHEDULE') and (var[4] == var[5]) and (var[4] == var[5]):
            return pd.Series([var[1], 'START', 'trigger'])
        
    return pd.Series([var[7], var[8], 'no'])
```

```python
df[['concept:name P', 'lifecycle:transition P', 'trigger_boy']] = df[['concept:name S1', 'concept:name S2','lifecycle:transition S1', 'lifecycle:transition S2', 
                           'case_id', 'case_id S1', 'case_id S2', 'concept:name', 'lifecycle:transition']].apply(absolute_concept, axis=1)
```
